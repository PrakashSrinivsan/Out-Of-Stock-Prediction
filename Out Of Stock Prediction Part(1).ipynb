> *The "Pampers Model to minimize the Out Of Stocks" project aims to address the challenge of optimizing product availability on store shelves during promotional events. The goal is to identify historical instances of Out Of Stocks during promotions and develop a predictive model to help distributors order the right volumes and avoid promotional Out Of Stock situations. By analyzing sales data, promotional records, and future promotional calendars, the project aims to provide a data-driven solution to maximize sales and prevent product unavailability during promotions.*

> **NOTE:** *Please note that due to the absence of information regarding the available stock volume and incomplete data spanning only a specific timeframe, it is not feasible to accurately predict and forecast the stock volume for this project. Therefore, we will approach this as an unsupervised problem. Using the available data, our focus will be on categorizing or labeling the records into "Instock" or "Out of Stock" categories based on the historical Out Of Stock instances during promotional events. This approach will provide valuable insights into understanding and addressing the issue of product availability during promotions, even without precise stock volume predictions.*

```python
#Pie Plot For Percentage of Items Sold with respect to Month

# Extract the month from the date column using the .dt accessor
sellout['month_inizio'] = sellout['Data_inizio'].dt.month

# Define a dictionary or a list to map month numbers to month names
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Count the number of items sold in each month
counts = sellout.groupby('month_inizio')['Pezzi'].sum()

# Calculate the percentage of products sold in each month
percentages = counts / counts.sum() * 100

# Sort the data by percentage in descending order
sorted_percentages = percentages.sort_values(ascending=True)

# Get the corresponding month names for each month number
month_labels = [month_names[month_num] for month_num in sorted_percentages.index]

# Define the base color and the number of steps
base_color = 'rgb(92, 129, 229)'
num_steps = len(sorted_percentages)

# Create a custom color scale with gradually decreasing intensity
color_scale = [base_color] + [f'rgb({int(92 * (1 - i / num_steps))}, {int(129 * (1 - i / num_steps))}, {int(229 * (1 - i / num_steps))})' for i in range(1, num_steps)]

# Create the pie chart using Plotly
fig = go.Figure(data=[go.Pie(labels=month_labels, values=sorted_percentages,
                             hovertemplate='Month: %{label}<br>Percentage: %{value:.1f}%')])

# Customize the colors using the color scale
fig.update_traces(marker=dict(colors=color_scale))

# Customize the width of the pie plot
fig.update_layout(width=800)

# Set the title
fig.update_layout(title='Percentage of Products Sold with Respect to Month')

# Display the interactive plot
pio.show(fig)
```
